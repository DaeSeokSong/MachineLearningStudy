```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # linestyle = 그려질 함수의 선의 모양(점선, 실선 등)
plt.xlabel("x") # xlabel = x축 이름
plt.ylabel("y") # ylabel = y축 이름
plt.title("sin & cos") # 해당 그래프 모형의 명칭
plt.legend() # 각 선이 어떤 함수인지 모형에 label을 띄워주는 메서드
plt.show()
```
